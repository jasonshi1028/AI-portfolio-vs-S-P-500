```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sp500 = "^GSPC"
ai_stocks = ["NVDA", "MSFT", "GOOGL", "META", "TSLA", "AMD", "AVGO"]
start_date = "2016-01-01"
end_date = "2026-03-01"
```

```python
sp500_data = yf.download(sp500, start=start_date, end=end_date)
ai_data = yf.download(ai_stocks, start=start_date, end=end_date)
```

```python
sp500_close = sp500_data["Close"]
ai_close = ai_data["Close"]
```

```python
print("S&P 500 data:")
print(sp500_close.head())
print("\nAI stock data:")
print(ai_close.head())
```

```python
sp500_normalized = sp500_close / sp500_close.iloc[0]
ai_normalized = ai_close / ai_close.iloc[0]
```

```python
plt.figure(figsize=(12, 6))

plt.plot(sp500_normalized, label="S&P 500", linewidth=3)

for stock in ai_normalized.columns:
    plt.plot(ai_normalized[stock], label=stock, alpha=0.7)

plt.title("AI Stocks vs S&P 500 (Normalized Prices)")
plt.xlabel("Date")
plt.ylabel("Normalized Price")
plt.legend()
plt.show()
```

```python
ai_portfolio = ai_normalized.mean(axis=1)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(ai_portfolio, label="AI Portfolio", linewidth=3)
plt.plot(sp500_normalized, label="S&P 500", linewidth=3)

plt.title("AI Portfolio vs S&P 500")
plt.xlabel("Date")
plt.ylabel("Normalized Price")
plt.legend()
plt.show()
```

```python
ai_total_return = (ai_portfolio.iloc[-1] / ai_portfolio.iloc[0] - 1) * 100
sp500_total_return = (sp500_normalized.iloc[-1] / sp500_normalized.iloc[0] - 1) * 100

print("AI Portfolio Total Return (%):", ai_total_return)
print("S&P 500 Total Return (%):", sp500_total_return)
print("Excess Return (%):", ai_total_return - sp500_total_return)
```

```python
ai_daily_returns = ai_portfolio.pct_change().dropna()
sp500_daily_returns = sp500_normalized.pct_change().dropna()
```

```python
ai_vol_daily = ai_daily_returns.std()
sp500_vol_daily = sp500_daily_returns.std()

ai_vol_annual = ai_vol_daily * (252 ** 0.5)
sp500_vol_annual = sp500_vol_daily * (252 ** 0.5)

print("AI Portfolio Annualized Volatility:", ai_vol_annual)
print("S&P 500 Annualized Volatility:", sp500_vol_annual)
```

```python
ai_return_decimal = ai_total_return / 100
sp500_return_decimal = sp500_total_return / 100

ai_risk_adjusted = ai_return_decimal / ai_vol_annual
sp500_risk_adjusted = sp500_return_decimal / sp500_vol_annual

print("AI Portfolio Risk-Adjusted Score:", ai_risk_adjusted)
print("S&P 500 Risk-Adjusted Score:", sp500_risk_adjusted)
```

```python
results = pd.DataFrame({
    "Metric": ["Total Return", "Annualized Volatility", "Risk-Adjusted Score"],
    "AI Portfolio": [ai_total_return, ai_vol_annual, ai_risk_adjusted],
    "S&P 500": [sp500_total_return, sp500_vol_annual, sp500_risk_adjusted]
})

print(results)
```

```python
ai_close.to_csv("ai_stock_prices.csv")
sp500_close.to_csv("sp500_prices.csv")
results.to_csv("research_results.csv", index=False)
```
